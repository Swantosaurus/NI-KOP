```kotlin
val df = DataFrame.read("result-20-91-Q.csv", ' ')//.filter { it["instance"] != "instance" }

df
```

```kotlin
val satisfied = df.get("satisfied").map { try {return@map (it as Int) } catch (e: Exception) {println(it); return@map 0}}.sum()



val all = df.get("runsPerInstance").map { (it as Int)}.sum()
println("satisfied" + (satisfied.toDouble() / all * 100).toString() + "%")


val satisfiedMaximum = df.get("satisfiendMaximum").map { (it as Int?) ?: 0}.sum()

println("satisfiedMaximum" + (satisfiedMaximum.toDouble() / all * 100).toString() + "%")


val steps = df.get("avrgIterations").values.map{it as Int}.average()
println("steps: $steps")

val relativeError = df.get("avrgDifference").values.map { it as Double }.average()

println("relativeError: $relativeError")


```

```kotlin
val p = File("p.txt").reader()
val pList = p.readLines().map { it.split(" ") }
val iterList = pList.map{ it.first().toInt() }
val probabilityList = pList.map { it[1].toDouble() }
//println(pList.max())
//print(pList.last())
letsPlot(mapOf("x" to iterList, "y" to probabilityList)) + geomLine {
    x = "x"
    y = "y"
} + xlab("iteration") + ylab("probability of accepting worse state")

```
