```kotlin
import java.io.File

val f = File("heuristicValue.txt").reader()
val liens = f.readLines().map{ it.toDouble() }
letsPlot(mapOf("x" to List<Int>(liens.size){it}, "y" to liens)) + geomLine {
    x = "x"
    y = "y"
}
```
